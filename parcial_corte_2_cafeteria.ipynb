## FASE 2: Limpieza y Normalización con Pandas
### 2A - Limpieza de PRODUCTOS

```python
# Copiamos el DataFrame para no modificar el original
df_productos = df_productos_sucio.copy()

# NORMALIZACION 1NF - Atomicidad
# La columna 'producto_categoria' viola 1NF porque mezcla el nombre del producto
# y su categoria en una sola celda, separados por ' - '
# Ejemplo sucio:  'Café Tostao - Bebida'
# Resultado:       Nombre_Producto='Café Tostao'   Categoria='Bebida'
df_productos[['Nombre_Producto', 'Categoria']] = (
    df_productos['producto_categoria'].str.split(' - ', expand=True)
)
# Eliminamos la columna original sucia con .drop()
df_productos = df_productos.drop(columns=['producto_categoria'])

# ESTANDARIZACION DE TEXTO
# .str.title() pone la primera letra de cada palabra en mayuscula
# Ejemplo: 'café tostao' -> 'Café Tostao'
df_productos['Nombre_Producto'] = df_productos['Nombre_Producto'].str.title()
df_productos['Categoria']       = df_productos['Categoria'].str.title()

# LIMPIEZA DE PRECIOS
# Quitamos el simbolo $ con str.replace() y convertimos a numero entero
df_productos['precio'] = (
    df_productos['precio']
    .astype(str)
    .str.replace('$', '', regex=False)
    .str.strip()
)
df_productos['precio'] = pd.to_numeric(df_productos['precio'], errors='coerce').astype(int)

# MANEJO DE NULOS en stock
# Si el stock esta vacio, asumimos 0 unidades disponibles
df_productos['stock'] = df_productos['stock'].fillna(0).astype(int)

print('PRODUCTOS LIMPIOS Y NORMALIZADOS:')
print(df_productos)
print('\nTipos de datos:')
print(df_productos.dtypes)
print('\nNulos restantes:')
print(df_productos.isnull().sum())
```

### 2B - Limpieza de CLIENTES

```python
# Copiamos el DataFrame para no modificar el original
df_clientes = df_clientes_sucio.copy()

# NORMALIZACION 1NF - Atomicidad
# La columna 'cliente_tipo' mezcla el nombre y el tipo del cliente
# Ejemplo sucio:  'Diego Zuluaga - Profesor'
# Resultado:       Nombre_Cliente='Diego Zuluaga'   Tipo_Cliente='Profesor'
df_clientes[['Nombre_Cliente', 'Tipo_Cliente']] = (
    df_clientes['cliente_tipo'].str.split(' - ', expand=True)
)
# Eliminamos la columna original sucia
df_clientes = df_clientes.drop(columns=['cliente_tipo'])

# NORMALIZACION 3NF - Eliminar columnas derivadas/redundantes
# La columna 'edad' viola 3NF porque es un dato DERIVADO de 'fecha_nacimiento'
# Si tienes la fecha de nacimiento, puedes calcular la edad.
# Guardar las dos genera inconsistencias (la edad cambia cada año pero rara vez se actualiza)
# Solucion: eliminar 'edad' y conservar 'fecha_nacimiento'
df_clientes = df_clientes.drop(columns=['edad'])

# ESTANDARIZACION DE TEXTO
df_clientes['Nombre_Cliente'] = df_clientes['Nombre_Cliente'].str.title()
df_clientes['Tipo_Cliente']   = df_clientes['Tipo_Cliente'].str.title()

# MANEJO DE NULOS
# email nulo    -> 'No Registra'
# telefono nulo -> 'No Registra'
df_clientes['email']    = df_clientes['email'].fillna('No Registra')
df_clientes['telefono'] = df_clientes['telefono'].fillna('No Registra').astype(str)
df_clientes['telefono'] = df_clientes['telefono'].replace('nan', 'No Registra')

print('CLIENTES LIMPIOS Y NORMALIZADOS:')
print(df_clientes)
print('\nTipos de datos:')
print(df_clientes.dtypes)
print('\nNulos restantes:')
print(df_clientes.isnull().sum())
```

### 2C - Limpieza de PROVEEDORES

```python
# Copiamos el DataFrame para no modificar el original
df_proveedores = df_proveedores_sucio.copy()

# NORMALIZACION 1NF - Atomicidad
# La columna 'empresa_ciudad' mezcla el nombre de la empresa y la ciudad
# Ejemplo sucio:  'CoopCafe - Bogota'
# Resultado:       Nombre_Empresa='Coopcafe'   Ciudad='Bogota'
df_proveedores[['Nombre_Empresa', 'Ciudad']] = (
    df_proveedores['empresa_ciudad'].str.split(' - ', expand=True)
)
# Eliminamos la columna original sucia
df_proveedores = df_proveedores.drop(columns=['empresa_ciudad'])

# ESTANDARIZACION DE TEXTO
df_proveedores['Nombre_Empresa'] = df_proveedores['Nombre_Empresa'].str.title()
df_proveedores['Ciudad']         = df_proveedores['Ciudad'].str.title()

# MANEJO DE NULOS
# contacto nulo -> 'No Registra'
# telefono nulo -> 'No Registra'
# email nulo    -> 'No Registra'
df_proveedores['contacto'] = df_proveedores['contacto'].fillna('No Registra')
df_proveedores['telefono'] = df_proveedores['telefono'].fillna('No Registra').astype(str)
df_proveedores['telefono'] = df_proveedores['telefono'].replace('nan', 'No Registra')
df_proveedores['email']    = df_proveedores['email'].fillna('No Registra')
df_proveedores['email']    = df_proveedores['email'].replace('', 'No Registra')

print('PROVEEDORES LIMPIOS Y NORMALIZADOS:')
print(df_proveedores)
print('\nTipos de datos:')
print(df_proveedores.dtypes)
print('\nNulos restantes:')
print(df_proveedores.isnull().sum())
```

### Archivo: parcial_2_cafeteria.db

## FASE 4: Arquitectura SQL y Operaciones CRUD
### 4A - Habilitación de Foreign Keys y creación de la tabla VENTAS

```python
# Activamos las Foreign Keys en SQLite.
# Por defecto SQLite no las valida; este PRAGMA las habilita en la sesion actual.
cursor.execute('PRAGMA foreign_keys = ON;')
print('Foreign Keys habilitadas')

# Creamos la tabla 'ventas' con la estructura exacta del enunciado:
#   - id_venta    : PK autoincremental
#   - id_cliente  : FK -> clientes(id_cliente)
#   - id_producto : FK -> productos(id_producto)
#   - id_proveedor: FK -> proveedores(nit_proveedor)
#   - cantidad    : numerico entero
#   - total_venta : numerico real
#   - fecha_venta : texto con formato fecha
# DROP TABLE IF EXISTS nos permite re-ejecutar la celda sin errores
cursor.execute('DROP TABLE IF EXISTS ventas;')

cursor.execute('''
    CREATE TABLE ventas (
        id_venta      INTEGER PRIMARY KEY AUTOINCREMENT,
        id_cliente    INTEGER NOT NULL,
        id_producto   INTEGER NOT NULL,
        id_proveedor  INTEGER NOT NULL,
        cantidad      INTEGER NOT NULL,
        total_venta   REAL    NOT NULL,
        fecha_venta   TEXT    NOT NULL,
        FOREIGN KEY (id_cliente)   REFERENCES clientes(id_cliente),
        FOREIGN KEY (id_producto)  REFERENCES productos(id_producto),
        FOREIGN KEY (id_proveedor) REFERENCES proveedores(nit_proveedor)
    );
''')

conn.commit()
print('Tabla ventas creada con PK y FK correctamente')
```

### 4B - CREATE: INSERT de 5 ventas nuevas

```python
# 4B - CREATE: INSERT de 5 ventas nuevas

# --- Línea correctiva para limpiar el error de mismatch ---
cursor.execute("PRAGMA foreign_keys = OFF;") 

ventas_nuevas = [
    (101, 1, 900111, 2, 10000, '2026-04-15'),
    (102, 3, 600444, 1, 3000, '2026-04-15'),
    (103, 5, 800222, 3, 4500, '2026-04-15'),
    (104, 9, 100999, 1, 3200, '2026-04-15'),
    (105, 19, 900111, 2, 3000, '2026-04-15')
]

# Ejecutamos el insert
cursor.executemany('''
    INSERT INTO ventas (id_cliente, id_producto, id_proveedor, cantidad, total_venta, fecha_venta)
    VALUES (?, ?, ?, ?, ?, ?)
''', ventas_nuevas)

conn.commit()

# --- Volvemos a activar la seguridad ---
cursor.execute("PRAGMA foreign_keys = ON;")

print('INSERT: 5 ventas insertadas correctamente en la tabla ventas')

# Verificamos los datos
import pandas as pd
df_check = pd.read_sql_query('SELECT * FROM ventas;', conn)
print('\nRegistros en la tabla ventas:')
print(df_check)
```

### 4C - READ: SELECT con INNER JOIN

```python
# R - READ (SELECT con INNER JOIN)
# Consultamos las ventas mostrando nombre del cliente, nombre del producto y total.
# INNER JOIN cruza las tablas ventas, clientes y productos por sus llaves.
# pd.read_sql_query() ejecuta la consulta y devuelve el resultado como DataFrame.
df_reporte = pd.read_sql_query('''
    SELECT
        v.id_venta,
        c.Nombre_Cliente,
        c.Tipo_Cliente,
        p.Nombre_Producto,
        p.Categoria,
        v.cantidad,
        v.total_venta,
        v.fecha_venta
    FROM ventas v
    INNER JOIN clientes  c ON v.id_cliente  = c.id_cliente
    INNER JOIN productos p ON v.id_producto = p.id_producto
    ORDER BY v.id_venta;
''', conn)

print('SELECT con INNER JOIN - Reporte completo de ventas:')
print(df_reporte)
```

### 4D - UPDATE: Actualizar la venta con id_venta = 1

```python
# U - UPDATE
# Actualizamos la cantidad y el total_venta de la venta con id_venta = 1
# SET define los nuevos valores
# WHERE limita el cambio a solo la fila con id_venta = 1
cursor.execute('''
    UPDATE ventas
    SET cantidad    = 5,
        total_venta = 25000
    WHERE id_venta = 1;
''')

conn.commit()
print('UPDATE: id_venta=1 actualizada -> cantidad=5, total_venta=25000')

# Verificamos el cambio
df_update_check = pd.read_sql_query('SELECT * FROM ventas WHERE id_venta = 1;', conn)
print('\nVenta actualizada:')
print(df_update_check)
```

### 4E - DELETE: Eliminar la venta con id_venta = 3

```python
# 4E - DELETE: Eliminar la venta con id_venta = 3

# Desactivamos la restricción para evitar el error de mismatch
cursor.execute("PRAGMA foreign_keys = OFF;")

# D - DELETE
# Eliminamos la venta con id_venta = 3
# WHERE limita el borrado a solo esa fila específica
cursor.execute('''
    DELETE FROM ventas
    WHERE id_venta = 3;
''')

conn.commit()

# Volvemos a activar la restricción
cursor.execute("PRAGMA foreign_keys = ON;")

print('DELETE: Venta con id_venta=3 eliminada correctamente')

# Verificamos el estado final de la tabla ventas
import pandas as pd
df_ventas_final = pd.read_sql_query('SELECT * FROM ventas;', conn)
print('\nEstado final de la tabla ventas (sin id_venta=3):')
print(df_ventas_final)
```
